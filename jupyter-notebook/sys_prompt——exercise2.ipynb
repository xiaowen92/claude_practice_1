### 3. Core Chat Function: `chat_log()`
```python
def chat_log(messages, system=None):
```

**Function Signature Analysis:**
- Uses `**params` dictionary to dynamically build API request parameters, providing flexibility
- `if system:` condition avoids sending `None` value to the API

**Why use `**params`:**
- Reduces code repetition (DRY principle) — no need for separate `if` statements for each parameter
- Easy to extend — can add new parameters (e.g., `temperature`, `top_p`) without modifying function signature
- More concise — single line `client.messages.create(**params)` instead of multiple conditional statements

**Example Extension:**
```python
def chat_log(messages, system=None, temperature=0.7, max_tokens=1000):
    params = {
        "model": model,
        "max_tokens": max_tokens,
        "temperature": temperature,
        "messages": messages
    }
    if system: 
        params["system"] = system
    return client.messages.create(**params).content[0].text
```

### 4. Example Usage
The third cell demonstrates the complete workflow: define system prompt, send user message, retrieve response.


````python
# load environment variables from .env file

from dotenv import load_dotenv

load_dotenv()

# load modules and set up client and model

from anthropic import Anthropic 
client = Anthropic()

model = "claude-sonnet-4-6"

````

````python
def add_user_message(messages, content):
    messages.append({
        "role": "user",
        "content": content
    })

def add_assistant_message(messages, content):
    messages.append({
        "role": "assistant",
        "content": content
    })

def chat_log (messages,system=None):
    # we do if condition to avoid sending system as None to the API, which will cause err
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }

    if system: 
        params["system"] = system

    messages = client.messages.create(**params)
    
    return messages.content[0].text

# ## previously what we have is 
# def chat_log (messages,system=None):
#     if system: 
#         messages = client.messages.create(
#             model=model,
#             max_tokens=1000,
#             messages=messages,
#             system=system
#         )
#     else:
#         messages = client.messages.create(
#             model=model,
#             max_tokens=1000,
#             messages=messages
````
