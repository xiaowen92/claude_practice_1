
**三个辅助函数（Helper Functions）：**

1. **`add_user_message(messages, content)`** — 将用户消息添加到消息列表
    - 参数：messages（消息列表）、content（用户输入内容）
    - 作用：以 `{"role": "user", "content": content}` 的格式追加到消息历史

2. **`add_assistant_message(messages, content)`** — 将助手响应添加到消息列表
    - 参数：messages（消息列表）、content（模型响应内容）
    - 作用：以 `{"role": "assistant", "content": content}` 的格式追加到消息历史

3. **`chat_log(messages)`** — 调用 Claude API 获取响应
    - 参数：messages（完整的消息历史）
    - 返回：模型生成的文本响应
    - 作用：将消息列表发送给 Claude Sonnet 4 模型，获取答案

**核心流程：**
这三个函数配合使用，实现了一个**多轮对话系统**（multi-turn conversation）：用户输入 → 模型响应 → 保存到历史 → 继续对话，形成连贯的上下文。

```python
# load environment variables from .env file

from dotenv import load_dotenv

load_dotenv()
```

```python
from anthropic import Anthropic 
client = Anthropic()

model = "claude-sonnet-4-6"

```

## Helper func to realize the multiple - converation available

```python
def add_user_message(messages, content):
    messages.append({
        "role": "user",
        "content": content
    })

def add_assistant_message(messages, content):
    messages.append({
        "role": "assistant",
        "content": content
    })

def chat_log (messages):
    messages = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages

    )
    return messages.content[0].text
```

```python
# define the initial messages
messages = []

# add the first user message

add_user_message(messages, "tell me how many planets in the solar system?")

print (messages)

# pass the messages to the chatlog function to get the response from the model

response = chat_log(messages)

print(response)

# we got message from the model, we can add it to the messages list as assistant message

add_assistant_message(messages, response)

print (messages)

# we can continue the conversation by adding more user messages and calling the chat_log function again

add_user_message(messages, "tell me more, such as which one is the largest?")

response = chat_log(messages)

print(response)

```
