# Multi-turn Conversational AI Assistant

This exercise implements an interactive chatbot based on the Anthropic Claude API.

## Workflow

1. **Initialization**: Load environment variables, create Anthropic client, set model to claude-sonnet-4-6
2. **Message Management**: Define two helper functions to manage conversation history
    - `add_user_message()` — Add user message to messages list
    - `add_assistant_message()` — Add AI response to messages list
3. **Chat Loop**: Continuously get user input until "exit" is entered
    - Each user input is added to message history
    - Call Claude API to generate response
    - AI response is added to message history
    - Display AI response and wait 0.5 seconds

## Key Features

- **Conversation Context Preservation**: messages list maintains complete conversation history, ensuring Claude can generate relevant responses based on previous exchanges
- **Continuous Interaction**: while loop enables multi-turn dialogue, allowing users to ask follow-up questions based on prior discussion

# 多轮对话 AI 助手 (Multi-turn Conversational AI Assistant)

这个练习实现了一个基于 Anthropic Claude API 的交互式聊天机器人。

## 工作流程

1. **初始化**: 加载环境变量、创建 Anthropic client、设置模型为 claude-sonnet-4-6
2. **消息管理**: 定义两个辅助函数用于管理对话历史
    - `add_user_message()` — 添加用户消息到 messages list
    - `add_assistant_message()` — 添加 AI 回复到 messages list
3. **聊天循环**: 持续获取用户输入直到输入 "exit"
    - 每次用户输入被添加到消息历史
    - 调用 Claude API 生成回复
    - AI 回复被添加到消息历史
    - 展示 AI 回复并等待 0.5 秒

## 关键特性

- **对话上下文保留**: messages list 维护完整的对话历史，确保 Claude 可以基于之前的对话内容生成相关回复
- **持续交互**: while 循环允许多轮对话，用户可以基于之前的讨论继续提问

## 演示结果

从上面的变量可以看到，用户先后提问了关于蓝鲸的四个问题（最大动物、饮食、潜水深度、更多信息），AI 都能基于对话历史提供连贯的回答。

```python
import time
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from anthropic import Anthropic 
client = Anthropic()

model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, content):
    messages.append({
        "role": "user",
        "content": content
    })

def add_assistant_message(messages, content):
    messages.append({
        "role": "assistant",
        "content": content
    })

def chat_log (messages):
    messages = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages

    )
    return messages.content[0].text
```

```python
user_input = ""
messages = []

while user_input.lower() != "exit": 

    user_input = input("User: ")

    print ('>', user_input)

    add_user_message(messages, user_input)

    assistant_response = chat_log(messages)

    add_assistant_message(messages, assistant_response)

    print ('Assistant:', assistant_response)

    time.sleep(0.5)
```
