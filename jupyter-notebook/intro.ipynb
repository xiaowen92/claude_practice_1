# A very simple one time conversation

```python
# load environment variables from .env file

from dotenv import load_dotenv

load_dotenv()
```

```python
from anthropic import Anthropic 
client = Anthropic()

model = "claude-sonnet-4-6"

```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "I need to learn some physics, tell me something about quantum mechanics."
        }
    ]
)

message
```
